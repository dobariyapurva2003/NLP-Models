```python
# =========================================================================
# CELL 1: Setup, Imports, Configuration, and Data Path Mapping
# =========================================================================

# Install necessary packages (Kaggle often has these pre-installed, but it's safe to include)
# !pip install transformers pandas numpy scikit-learn matplotlib seaborn torch
# !pip install gensim --upgrade

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import DataLoader, Dataset, TensorDataset
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, f1_score, precision_score, recall_score, confusion_matrix
import time
import re
import os
import random
from transformers import BertTokenizer, BertForSequenceClassification, get_linear_schedule_with_warmup
from tqdm.auto import tqdm
from gensim.models import KeyedVectors
from collections import Counter
# Set CUDA_LAUNCH_BLOCKING to 1 for synchronous error reporting/debugging
os.environ['CUDA_LAUNCH_BLOCKING'] = '1'

# ===========================
# Configuration and Reproducibility
# ===========================

SEED_VALUE = 42
RNN_LSTM_GRU_LR = 1e-5
BERT_LR = 1e-5
RNN_LSTM_GRU_EPOCHS = 15
BERT_EPOCHS = 4 # Use 4 epochs for BERT (the standard fine-tuning range)
MAX_SEQ_LENGTH = 256
RNN_LSTM_GRU_BATCH_SIZE = 64
BERT_BATCH_SIZE = 32 # Optimized for GPU training of BERT
MODEL_SAVE_DIR = '/kaggle/working'
os.makedirs(MODEL_SAVE_DIR, exist_ok=True)
print(f"Models will be saved to: {MODEL_SAVE_DIR}")

def set_all_seeds(seed_value):
    """Sets seeds for reproducibility across runs."""
    random.seed(seed_value)
    np.random.seed(seed_value)
    torch.manual_seed(seed_value)
    if torch.cuda.is_available():
        torch.cuda.manual_seed_all(seed_value)
        torch.backends.cudnn.deterministic = True
        torch.backends.cudnn.benchmark = False

set_all_seeds(SEED_VALUE)

device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print(f"Using device: {device}")

# =========================================================================
# !! IMPORTANT: UPDATE THESE PATHS for your Kaggle Environment !!
# =========================================================================
# On Kaggle, datasets are usually mounted under the '../input/' directory.
# Adjust the paths based on the names you gave your dataset folders.

# 1. IMDB Dataset (CSV file)
DATA_PATH = '../input/imdb-dataset-of-50k-movie-reviews/IMDB Dataset.csv'

# 2. GloVe Embeddings (txt file)
EMBEDDING_PATH = '../input/glove6b100dtxt/glove.6B.100d.txt'

# 3. BERT Model Cache (Folder, used for offline/faster loading)
BERT_CACHE_DIR = '../input/bertbaseuncased/bert-base-uncased'

# =========================================================================
# CELL 2: Data Preprocessing
# =========================================================================

def clean_text(text):
    """Basic text cleaning for IMDB reviews."""
    text = re.sub(r'<[^>]*>', '', text)  # Remove HTML tags
    text = re.sub(r'[^a-zA-Z\s]', '', text.lower()) # Keep only letters and spaces, lowercase
    return text.strip()

def load_and_preprocess_imdb_data(file_path, test_size=0.2, val_size=0.1):
    """Loads IMDB data, cleans, converts sentiment, and splits into train/val/test."""
    print("Loading and preprocessing data...")
    try:
        df = pd.read_csv(file_path)
    except FileNotFoundError:
        print(f"Error: File not found at {file_path}. Please update DATA_PATH.")
        return None, None, None

    df['review'] = df['review'].apply(clean_text)
    # Convert 'sentiment' ('positive'/'negative') to numerical (1/0)
    df['sentiment'] = df['sentiment'].apply(lambda x: 1 if x == 'positive' else 0)

    X = df['review'].tolist()
    y = df['sentiment'].values

    # Split into temporary training (80%) and test (20%) sets
    X_train_val, X_test, y_train_val, y_test = train_test_split(
        X, y, test_size=test_size, random_state=SEED_VALUE, stratify=y
    )

    # Split temporary training set into final training and validation (12.5% of temp = 10% of total)
    X_train, X_val, y_train, y_val = train_test_split(
        X_train_val, y_train_val, test_size=val_size / (1 - test_size), random_state=SEED_VALUE, stratify=y_train_val
    )

    print(f"Total samples: {len(df)}, Train: {len(X_train)}, Val: {len(X_val)}, Test: {len(X_test)}")

    return (X_train, y_train), (X_val, y_val), (X_test, y_test)

# Load and split the data
train_data_raw, val_data_raw, test_data_raw = load_and_preprocess_imdb_data(DATA_PATH)

# =========================================================================
# CELL 3: GloVe Preprocessing (for RNN, LSTM, GRU)
# =========================================================================

def tokenize_and_build_vocab(data_raw):
    """Tokenizes data and builds a vocabulary based on training data."""
    # data_raw is expected to be a tuple of (X_train, y_train)
    # X_train is the first element of that tuple, data_raw[0]
    X_train = data_raw[0] # CORRECTED: Access the list of texts (X_train)
    
    tokenized_texts = [text.split() for text in X_train]
    word_counts = Counter(token for tokens in tokenized_texts for token in tokens)
    
    # Create vocabulary: 0 for padding, 1 for OOV (unknown)
    vocab = {'<pad>': 0, '<unk>': 1}
    for word, count in word_counts.most_common():
        vocab[word] = len(vocab)
        
    return vocab, word_counts

def load_glove_embeddings(vocab, embedding_path, embedding_dim=100):
    print(f"Loading GloVe embeddings from {embedding_path}...")
    try:
        # Load GloVe vectors
        word_vectors = KeyedVectors.load_word2vec_format(embedding_path, binary=False, no_header=True)
    except FileNotFoundError:
        print("Error: GloVe file not found. Using random embeddings.")
        return torch.randn(len(vocab), embedding_dim)

    embedding_matrix = np.zeros((len(vocab), embedding_dim))
    for word, i in tqdm(vocab.items(), desc="Building Embedding Matrix"):
        if word in word_vectors:
            embedding_matrix[i] = word_vectors[word]
        elif word != '<pad>':
             # For <unk> or OOV words, use random uniform initialization
             embedding_matrix[i] = np.random.uniform(-0.1, 0.1, embedding_dim)

    return torch.tensor(embedding_matrix, dtype=torch.float32)

def sequence_to_tensor(texts, vocab, max_seq_length=MAX_SEQ_LENGTH):
    """Converts tokenized sequences to padded tensor of indices."""
    tokenized_texts = [text.split() for text in texts]
    sequences = []
    for tokens in tokenized_texts:
        indices = [vocab.get(token, vocab['<unk>']) for token in tokens]
        if len(indices) > max_seq_length:
            indices = indices[:max_seq_length] # Truncate
        elif len(indices) < max_seq_length:
            indices += [vocab['<pad>']] * (max_seq_length - len(indices)) # Pad
        sequences.append(indices)
        
    return torch.tensor(sequences, dtype=torch.long)

# 1. Tokenize and Build Vocab
vocab, _ = tokenize_and_build_vocab(train_data_raw)
VOCAB_SIZE = len(vocab)
EMBEDDING_DIM = 100

# 2. Load Embeddings
glove_embedding_matrix = load_glove_embeddings(vocab, EMBEDDING_PATH, EMBEDDING_DIM)

# 3. Convert Data to Tensors and create DataLoaders
X_train_glove = sequence_to_tensor(train_data_raw[0], vocab)
y_train_glove = torch.tensor(train_data_raw[1], dtype=torch.long)
X_val_glove = sequence_to_tensor(val_data_raw[0], vocab)
y_val_glove = torch.tensor(val_data_raw[1], dtype=torch.long)
X_test_glove = sequence_to_tensor(test_data_raw[0], vocab)
y_test_glove = torch.tensor(test_data_raw[1], dtype=torch.long)

train_dataset_glove = TensorDataset(X_train_glove, y_train_glove)
val_dataset_glove = TensorDataset(X_val_glove, y_val_glove)
test_dataset_glove = TensorDataset(X_test_glove, y_test_glove)

train_loader_glove = DataLoader(train_dataset_glove, batch_size=RNN_LSTM_GRU_BATCH_SIZE, shuffle=True)
val_loader_glove = DataLoader(val_dataset_glove, batch_size=RNN_LSTM_GRU_BATCH_SIZE, shuffle=False)
test_loader_glove = DataLoader(test_dataset_glove, batch_size=RNN_LSTM_GRU_BATCH_SIZE, shuffle=False)

print("GloVe Preprocessing Complete.")

# =========================================================================
# CELL 4: BERT Preprocessing
# =========================================================================

class BertSentimentDataset(Dataset):
    def __init__(self, texts, labels, tokenizer, max_seq_length=MAX_SEQ_LENGTH):
        self.texts = texts
        self.labels = labels
        self.tokenizer = tokenizer
        self.max_seq_length = max_seq_length

    def __len__(self):
        return len(self.texts)

    def __getitem__(self, idx):
        text = self.texts[idx]
        label = self.labels[idx]

        encoding = self.tokenizer.encode_plus(
            text,
            add_special_tokens=True,
            max_length=self.max_seq_length,
            padding='max_length',
            truncation=True,
            return_attention_mask=True,
            return_tensors='pt',
        )

        return {
            'input_ids': encoding['input_ids'].flatten(),
            'attention_mask': encoding['attention_mask'].flatten(),
            'labels': torch.tensor(label, dtype=torch.long)
        }


def prepare_bert_dataloader(data_raw):
    # Using the local BERT files for faster loading/offline capability
    bert_vocab_file = os.path.join(BERT_CACHE_DIR, 'vocab.txt')
    # Note: Kaggle notebooks often require the full path to the vocab file when loading offline.
    tokenizer = BertTokenizer.from_pretrained(bert_vocab_file, local_files_only=True) 

    # CORRECTED: data_raw is already (X_train, y_train), so we unpack the tuple directly.
    X_train, y_train = data_raw
    # CORRECTED: val_data_raw is a global variable (X_val, y_val), so we unpack the tuple directly.
    X_val, y_val = val_data_raw
    # CORRECTED: test_data_raw is a global variable (X_test, y_test), so we unpack the tuple directly.
    X_test, y_test = test_data_raw

    train_dataset = BertSentimentDataset(X_train, y_train, tokenizer)
    val_dataset = BertSentimentDataset(X_val, y_val, tokenizer)
    test_dataset = BertSentimentDataset(X_test, y_test, tokenizer)

    train_loader = DataLoader(train_dataset, batch_size=BERT_BATCH_SIZE, shuffle=True)
    val_loader = DataLoader(val_dataset, batch_size=BERT_BATCH_SIZE, shuffle=False)
    test_loader = DataLoader(test_dataset, batch_size=BERT_BATCH_SIZE, shuffle=False)

    print("BERT Preprocessing Complete.")
    return train_loader, val_loader, test_loader

train_loader_bert, val_loader_bert, test_loader_bert = prepare_bert_dataloader(train_data_raw)


# =========================================================================
# CELL 5: Model Definitions
# =========================================================================

# BERT Model
class BERTSentimentClassifier(nn.Module):
    def __init__(self, num_classes=2):
        super(BERTSentimentClassifier, self).__init__()
        # Load BERT model from local cache
        self.bert = BertForSequenceClassification.from_pretrained(
            BERT_CACHE_DIR, 
            num_labels=num_classes, 
            local_files_only=True
        )

    def forward(self, input_ids, attention_mask, labels=None):
        outputs = self.bert(
            input_ids=input_ids, 
            attention_mask=attention_mask, 
            labels=labels # Pass labels to the underlying BERT model
        )
        
        # The BERT model returns an object with attributes like .loss and .logits
        # We must return this outputs object so that the train_bert function can access outputs.loss
        return outputs

# Base RNN/LSTM/GRU Model
class BaseRNNModel(nn.Module):
    def __init__(self, embedding_matrix, hidden_dim, num_layers, dropout, bidirectional, model_type):
        super(BaseRNNModel, self).__init__()
        
        # Initialize embedding layer with pre-trained GloVe vectors (freeze weights)
        self.embedding = nn.Embedding.from_pretrained(embedding_matrix, freeze=True, padding_idx=0)
        self.hidden_dim = hidden_dim
        self.num_directions = 2 if bidirectional else 1
        
        # Select RNN layer type
        rnn_params = dict(
            input_size=EMBEDDING_DIM, hidden_size=hidden_dim, num_layers=num_layers,
            batch_first=True, dropout=dropout, bidirectional=bidirectional
        )
        if model_type == 'RNN':
            self.rnn_layer = nn.RNN(**rnn_params)
        elif model_type == 'LSTM':
            self.rnn_layer = nn.LSTM(**rnn_params)
        elif model_type == 'GRU':
            self.rnn_layer = nn.GRU(**rnn_params)
        else:
            raise ValueError("Invalid model_type")
            
        # Classifier layer
        self.fc = nn.Linear(hidden_dim * self.num_directions, 2)
        self.dropout = nn.Dropout(dropout)

    def forward(self, text):
        embedded = self.dropout(self.embedding(text))
        output, hidden = self.rnn_layer(embedded) 
        
        # Use the output of the final time step of the last layer
        if self.num_directions == 2:
            # Concatenate the last time step of the forward sequence (output[:, -1, :hidden_dim]) 
            # and the first time step of the backward sequence (output[:, 0, hidden_dim:])
            hidden_concat = torch.cat((output[:, -1, :self.hidden_dim], output[:, 0, self.hidden_dim:]), dim=1)
        else:
            hidden_concat = output[:, -1, :]

        logits = self.fc(self.dropout(hidden_concat))
        return logits

def RNNSentimentClassifier(embedding_matrix, hidden_dim=256, num_layers=2, dropout=0.5, bidirectional=True):
    return BaseRNNModel(embedding_matrix, hidden_dim, num_layers, dropout, bidirectional, 'RNN')

def LSTMSentimentClassifier(embedding_matrix, hidden_dim=256, num_layers=2, dropout=0.5, bidirectional=True):
    return BaseRNNModel(embedding_matrix, hidden_dim, num_layers, dropout, bidirectional, 'LSTM')

def GRUSentimentClassifier(embedding_matrix, hidden_dim=256, num_layers=2, dropout=0.5, bidirectional=True):
    return BaseRNNModel(embedding_matrix, hidden_dim, num_layers, dropout, bidirectional, 'GRU')

print("Model Definitions Complete.")

# =========================================================================
# CELL 6: Training and Evaluation Functions
# =========================================================================

def evaluate_model_metrics(all_labels, all_preds, total_samples, inference_time):
    """Calculates all required metrics from predictions and time."""
    accuracy = accuracy_score(all_labels, all_preds)
    f1 = f1_score(all_labels, all_preds, average='binary')
    precision = precision_score(all_labels, all_preds, average='binary')
    recall = recall_score(all_labels, all_preds, average='binary')
    cm = confusion_matrix(all_labels, all_preds)
    inference_speed_sps = total_samples / inference_time if inference_time > 0 else 0
    return accuracy, f1, precision, recall, cm, inference_speed_sps

def train_model(model, train_loader, val_loader, optimizer, criterion, epochs, model_name):
    """Training function for RNN/LSTM/GRU models."""
    model.to(device)
    history = {'train_loss': [], 'val_loss': [], 'train_acc': [], 'val_acc': []}
    best_val_acc = -1.0 # Initialize best accuracy
    
    start_time = time.time()
    
    for epoch in range(epochs):
        model.train()
        total_loss, total_correct, total_samples = 0, 0, 0
        
        # Training phase
        for inputs, labels in tqdm(train_loader, desc=f"Epoch {epoch+1}/{epochs} (Train)"):
            inputs, labels = inputs.to(device), labels.to(device)
            
            optimizer.zero_grad()
            outputs = model(inputs)
            loss = criterion(outputs, labels)
            loss.backward()
            optimizer.step()
            
            total_loss += loss.item() * inputs.size(0)
            _, predicted = torch.max(outputs, 1)
            total_correct += (predicted == labels).sum().item()
            total_samples += inputs.size(0)
            
        avg_train_loss = total_loss / total_samples
        train_accuracy = total_correct / total_samples
        
        # Validation phase
        model.eval()
        val_loss, val_correct, val_samples = 0, 0, 0
        with torch.no_grad():
            for inputs, labels in val_loader:
                inputs, labels = inputs.to(device), labels.to(device)
                outputs = model(inputs)
                loss = criterion(outputs, labels)
                
                val_loss += loss.item() * inputs.size(0)
                _, predicted = torch.max(outputs, 1)
                val_correct += (predicted == labels).sum().item()
                val_samples += inputs.size(0)
                
        avg_val_loss = val_loss / val_samples
        val_accuracy = val_correct / val_samples
        best_val_acc = save_best_checkpoint(model, model_name, val_accuracy, best_val_acc)
        
        # Store history
        history['train_loss'].append(avg_train_loss)
        history['val_loss'].append(avg_val_loss)
        history['train_acc'].append(train_accuracy)
        history['val_acc'].append(val_accuracy)
        
        print(f'{model_name} Epoch {epoch+1}: Train Loss: {avg_train_loss:.4f}, Val Acc: {val_accuracy:.4f}')

    end_time = time.time()
    total_training_time = end_time - start_time
    print(f"\n{model_name} Total Training Time: {total_training_time:.2f} seconds")
    
    # After training, load the best model for final evaluation
    path = os.path.join(MODEL_SAVE_DIR, f'{model_name}_best_model.pth')
    model.load_state_dict(torch.load(path))
    print(f"Loaded best checkpoint for {model_name} for final evaluation.")
    
    return model, total_training_time, history
    
    return model, total_training_time, history

def evaluate_model(model, test_loader, criterion):
    """Evaluation function for RNN/LSTM/GRU models (includes inference speed)."""
    model.eval()
    all_labels, all_preds = [], []
    total_samples = 0
    
    start_inference_time = time.time()
    
    with torch.no_grad():
        for inputs, labels in tqdm(test_loader, desc="Evaluating Model"):
            inputs, labels = inputs.to(device), labels.to(device)
            outputs = model(inputs)
            
            total_samples += inputs.size(0)
            _, predicted = torch.max(outputs, 1)
            all_labels.extend(labels.cpu().numpy())
            all_preds.extend(predicted.cpu().numpy())
            
    end_inference_time = time.time()
    inference_time = end_inference_time - start_inference_time
    
    return evaluate_model_metrics(all_labels, all_preds, total_samples, inference_time)


def train_bert(model, train_loader, val_loader, optimizer, scheduler, epochs, model_name="BERT"):
    """Training function for BERT model."""
    model.to(device)
    history = {'train_loss': [], 'val_loss': [], 'train_acc': [], 'val_acc': []}
    best_val_acc = -1.0 # Initialize best accuracy
    
    start_time = time.time()
    
    for epoch in range(epochs):
        model.train()
        total_loss, total_correct, total_samples = 0, 0, 0
        
        # Training phase
        for batch in tqdm(train_loader, desc=f"Epoch {epoch+1}/{epochs} (Train)"):
            input_ids = batch['input_ids'].to(device)
            attention_mask = batch['attention_mask'].to(device)
            labels = batch['labels'].to(device)
            
            optimizer.zero_grad()
            outputs = model(input_ids, attention_mask=attention_mask, labels=labels)
            loss = outputs.loss
            logits = outputs.logits
            loss.backward()
            
            torch.nn.utils.clip_grad_norm_(model.parameters(), 1.0) # Gradient clipping
            
            optimizer.step()
            scheduler.step()
            
            total_loss += loss.item()
            _, predicted = torch.max(logits, 1)
            total_correct += (predicted == labels).sum().item()
            total_samples += labels.size(0)
            
        avg_train_loss = total_loss / len(train_loader)
        train_accuracy = total_correct / total_samples
        
        # Validation phase
        model.eval()
        val_loss, val_correct, val_samples = 0, 0, 0
        with torch.no_grad():
            for batch in val_loader:
                input_ids = batch['input_ids'].to(device)
                attention_mask = batch['attention_mask'].to(device)
                labels = batch['labels'].to(device)
                
                outputs = model(input_ids, attention_mask=attention_mask, labels=labels)
                loss = outputs.loss
                logits = outputs.logits
                
                val_loss += loss.item()
                _, predicted = torch.max(logits, 1)
                val_correct += (predicted == labels).sum().item()
                val_samples += labels.size(0)
                
        avg_val_loss = val_loss / len(val_loader)
        val_accuracy = val_correct / val_samples
        best_val_acc = save_best_checkpoint(model, model_name, val_accuracy, best_val_acc)
        
        # Store history
        history['train_loss'].append(avg_train_loss)
        history['val_loss'].append(avg_val_loss)
        history['train_acc'].append(train_accuracy)
        history['val_acc'].append(val_accuracy)
        
        print(f'{model_name} Epoch {epoch+1}: Train Loss: {avg_train_loss:.4f}, Val Acc: {val_accuracy:.4f}')

    end_time = time.time()
    total_training_time = end_time - start_time
    print(f"\n{model_name} Total Training Time: {total_training_time:.2f} seconds")
    # After training, load the best model for final evaluation
    path = os.path.join(MODEL_SAVE_DIR, f'{model_name}_best_model.pth')
    model.load_state_dict(torch.load(path))
    print(f"Loaded best checkpoint for {model_name} for final evaluation.")
    
    return model, total_training_time, history

def evaluate_bert(model, test_loader):
    """Evaluation function for BERT (includes inference speed)."""
    model.eval()
    all_labels, all_preds = [], []
    total_samples = 0
    
    start_inference_time = time.time()
    
    with torch.no_grad():
        for batch in tqdm(test_loader, desc="Evaluating BERT Model"):
            input_ids = batch['input_ids'].to(device)
            attention_mask = batch['attention_mask'].to(device)
            labels = batch['labels'].to(device)
            
            outputs = model(input_ids=input_ids, attention_mask=attention_mask)
            logits = outputs.logits
            
            total_samples += labels.size(0)
            
            _, predicted = torch.max(logits, 1)
            all_labels.extend(labels.cpu().numpy())
            all_preds.extend(predicted.cpu().numpy())
            
    end_inference_time = time.time()
    inference_time = end_inference_time - start_inference_time
    
    return evaluate_model_metrics(all_labels, all_preds, total_samples, inference_time)
    
def save_best_checkpoint(model, model_name, val_acc, best_val_acc, save_dir=MODEL_SAVE_DIR):
    """Saves the trained model state dictionary if current val_acc is better than best_val_acc."""
    if val_acc > best_val_acc:
        path = os.path.join(save_dir, f'{model_name}_best_model.pth')
        torch.save(model.state_dict(), path)
        print(f"--> NEW BEST: {model_name} model saved with Val Acc: {val_acc:.4f}")
        return val_acc
    return best_val_acc

# =========================================================================
# CELL 7: Model Execution and Comparison
# =========================================================================

results = {}
criterion = nn.CrossEntropyLoss()

# --- RNN Model ---
print("\n--- Training RNN ---")
rnn_model = RNNSentimentClassifier(glove_embedding_matrix).to(device)
rnn_optimizer = optim.Adam(rnn_model.parameters(), lr=RNN_LSTM_GRU_LR)
rnn_model_trained, rnn_train_time, rnn_history = train_model(rnn_model, train_loader_glove, val_loader_glove, rnn_optimizer, criterion, RNN_LSTM_GRU_EPOCHS , "RNN")
rnn_acc, rnn_f1, rnn_prec, rnn_rec, rnn_cm, rnn_sps = evaluate_model(rnn_model_trained, test_loader_glove, criterion)
results['RNN'] = {'Accuracy': rnn_acc, 'F1-Score': rnn_f1, 'Confusion Matrix': rnn_cm, 'Training Time (s)': rnn_train_time, 'Inference Speed (SPS)': rnn_sps, 'History': rnn_history}


# --- LSTM Model ---
print("\n--- Training LSTM ---")
lstm_model = LSTMSentimentClassifier(glove_embedding_matrix).to(device)
lstm_optimizer = optim.Adam(lstm_model.parameters(), lr=RNN_LSTM_GRU_LR)
lstm_model_trained, lstm_train_time, lstm_history = train_model(lstm_model, train_loader_glove, val_loader_glove, lstm_optimizer, criterion, RNN_LSTM_GRU_EPOCHS, "LSTM")
lstm_acc, lstm_f1, lstm_prec, lstm_rec, lstm_cm, lstm_sps = evaluate_model(lstm_model_trained, test_loader_glove, criterion)
results['LSTM'] = {'Accuracy': lstm_acc, 'F1-Score': lstm_f1, 'Confusion Matrix': lstm_cm, 'Training Time (s)': lstm_train_time, 'Inference Speed (SPS)': lstm_sps, 'History': lstm_history}


# --- GRU Model ---
print("\n--- Training GRU ---")
gru_model = GRUSentimentClassifier(glove_embedding_matrix).to(device)
gru_optimizer = optim.Adam(gru_model.parameters(), lr=RNN_LSTM_GRU_LR)
gru_model_trained, gru_train_time, gru_history = train_model(gru_model, train_loader_glove, val_loader_glove, gru_optimizer, criterion, RNN_LSTM_GRU_EPOCHS, "GRU")
gru_acc, gru_f1, gru_prec, gru_rec, gru_cm, gru_sps = evaluate_model(gru_model_trained, test_loader_glove, criterion)
results['GRU'] = {'Accuracy': gru_acc, 'F1-Score': gru_f1, 'Confusion Matrix': gru_cm, 'Training Time (s)': gru_train_time, 'Inference Speed (SPS)': gru_sps, 'History': gru_history}


# --- BERT Model ---
print("\n--- Training BERT ---")
bert_model = BERTSentimentClassifier(num_classes=2).to(device)
total_steps = len(train_loader_bert) * BERT_EPOCHS
bert_optimizer = optim.AdamW(bert_model.parameters(), lr=BERT_LR, eps=1e-8)
bert_scheduler = get_linear_schedule_with_warmup(bert_optimizer, num_warmup_steps=0, num_training_steps=total_steps)

bert_model_trained, bert_train_time, bert_history = train_bert(bert_model, train_loader_bert, val_loader_bert, bert_optimizer, bert_scheduler, BERT_EPOCHS, "BERT")
bert_acc, bert_f1, bert_prec, bert_rec, bert_cm, bert_sps = evaluate_bert(bert_model_trained, test_loader_bert)
results['BERT'] = {'Accuracy': bert_acc, 'F1-Score': bert_f1, 'Confusion Matrix': bert_cm, 'Training Time (s)': bert_train_time, 'Inference Speed (SPS)': bert_sps, 'History': bert_history}


# =========================================================================
# CELL 8: Results Visualization and Analysis
# =========================================================================

def plot_learning_curves(history, model_name):
    """Plots the training and validation loss and accuracy over epochs."""
    epochs = range(1, len(history['train_loss']) + 1)
    fig, ax = plt.subplots(1, 2, figsize=(14, 5))
    fig.suptitle(f'{model_name} Learning Curves', fontsize=16)

    # Plot Loss
    ax[0].plot(epochs, history['train_loss'], 'bo-', label='Training Loss')
    ax[0].plot(epochs, history['val_loss'], 'ro-', label='Validation Loss')
    ax[0].set_title('Loss vs. Epochs')
    ax[0].set_xlabel('Epoch')
    ax[0].set_ylabel('Loss')
    ax[0].legend()
    ax[0].grid(True)

    # Plot Accuracy
    ax[1].plot(epochs, history['train_acc'], 'bo-', label='Training Accuracy')
    ax[1].plot(epochs, history['val_acc'], 'ro-', label='Validation Accuracy')
    ax[1].set_title('Accuracy vs. Epochs')
    ax[1].set_xlabel('Epoch')
    ax[1].set_ylabel('Accuracy')
    ax[1].legend()
    ax[1].grid(True)
    
    plt.tight_layout(rect=[0, 0.03, 1, 0.95])
    plt.show()
    
def plot_confusion_matrix(cm, model_name):
    """Plots a confusion matrix for the model."""
    plt.figure(figsize=(6, 5))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', cbar=False, 
                xticklabels=['Negative (0)', 'Positive (1)'], yticklabels=['Negative (0)', 'Positive (1)'])
    plt.title(f'{model_name} Confusion Matrix')
    plt.xlabel('Predicted Labels')
    plt.ylabel('True Labels')
    plt.show()

def plot_comparison_metrics(results):
    """Plots performance and efficiency metrics for all models."""
    df = pd.DataFrame(results).T.drop(columns=['Confusion Matrix', 'History'])
    
    # 1. Performance Metrics (Accuracy, F1-Score)
    performance_df = df[['Accuracy', 'F1-Score']]
    performance_df.plot(kind='bar', figsize=(10, 6))
    plt.title('Performance Comparison (Accuracy & F1-Score)')
    plt.xlabel('Models')
    plt.ylabel('Score')
    plt.xticks(rotation=0)
    plt.grid(axis='y', linestyle='--')
    plt.show()
    
    # 2. Efficiency Metrics (Training Time)
    plt.figure(figsize=(8, 5))
    df['Training Time (s)'].plot(kind='bar', color='skyblue')
    plt.title('Training Time Comparison')
    plt.xlabel('Models')
    plt.ylabel('Time (seconds)')
    plt.xticks(rotation=0)
    plt.grid(axis='y', linestyle='--')
    plt.show()
    
    # 3. Efficiency Metrics (Inference Speed)
    plt.figure(figsize=(8, 5))
    df['Inference Speed (SPS)'].plot(kind='bar', color='lightcoral')
    plt.title('Inference Speed Comparison (Samples Per Second)')
    plt.xlabel('Models')
    plt.ylabel('Speed (SPS)')
    plt.xticks(rotation=0)
    plt.grid(axis='y', linestyle='--')
    plt.show()

# Display the results table
results_df = pd.DataFrame(results).T.drop(columns=['Confusion Matrix', 'History']).round(4)
print("\n=======================================================")
print("              FINAL MODEL COMPARISON TABLE           ")
print("=======================================================")
print(results_df)
print("=======================================================")

# Plotting
plot_comparison_metrics(results)

# Detailed analysis for each model
for model_name, data in results.items():
    print(f"\n--- Analysis for {model_name} ---")
    plot_learning_curves(data['History'], model_name)
    plot_confusion_matrix(data['Confusion Matrix'], model_name)
    
print("Analysis Complete.")
```

```python
# =========================================================================
# CELL 9: Prediction Setup & Model Loading
# =========================================================================
import os
import torch.nn.functional as F
from transformers import BertTokenizer
# Ensure pandas is imported for the final table display
import pandas as pd 

# --- 1. Initialize BERT Tokenizer ---
# Re-initialize BERT tokenizer using the local cache files
BERT_CACHE_DIR = '/kaggle/input/bertbaseuncased/bert-base-uncased'
bert_vocab_file = os.path.join(BERT_CACHE_DIR, 'vocab.txt')

try:
    live_tokenizer = BertTokenizer.from_pretrained(bert_vocab_file, local_files_only=True)
    print("BERT Tokenizer loaded for live inference.")
except Exception as e:
    print(f"ERROR: Could not load BERT Tokenizer. Check BERT_CACHE_DIR path. Error: {e}")
    live_tokenizer = None

# --- 2. Load Saved Model Weights ---
MODEL_SAVE_DIR = '/kaggle/working'
loaded_models = {}

# Helper to load models (assuming model classes from Cell 5 are in scope)
def load_model_for_inference(name, model_factory, is_bert=False):
    try:
        path = os.path.join(MODEL_SAVE_DIR, f'{name}_best_model.pth')
        
        # Instantiate the model architecture
        if is_bert:
             # BERTSentimentClassifier uses a default constructor
            model = model_factory(num_classes=2).to(device)
        else:
            # RNN/LSTM/GRU need the GloVe matrix
            model = model_factory(glove_embedding_matrix).to(device) 

        # Load the saved state dictionary
        # Note: strict=False is used for BERT to avoid errors if any unused heads are missing from the saved state
        model.load_state_dict(torch.load(path, map_location=device), strict=False)
        model.eval() # Set to evaluation mode (crucial for prediction)
        loaded_models[name] = model
        print(f"Loaded {name} model.")
    except Exception as e:
        print(f"Error loading {name} model: {e}")

# Load all four models
load_model_for_inference("RNN", RNNSentimentClassifier)
load_model_for_inference("LSTM", LSTMSentimentClassifier)
load_model_for_inference("GRU", GRUSentimentClassifier)
load_model_for_inference("BERT", BERTSentimentClassifier, is_bert=True)

print("\nAll available models for prediction:", list(loaded_models.keys()))
```

```python
# =========================================================================
# CELL 10: Unified Prediction Function
# =========================================================================

def predict_sentiment_live(review_text, model_name):
    """
    Predicts the sentiment and confidence score for a single raw review.
    
    Args:
        review_text (str): The raw movie review text.
        model_name (str): 'RNN', 'LSTM', 'GRU', or 'BERT'.
        
    Returns:
        tuple: (predicted_sentiment_label, confidence_score_percent)
    """
    model = loaded_models.get(model_name)
    if model is None:
        return "Model not loaded", 0.0

    # 1. Preprocess the Text
    cleaned_text = clean_text(review_text)
    
    with torch.no_grad():
        if model_name in ['RNN', 'LSTM', 'GRU']:
            # --- GloVe-based Preprocessing (Uses clean_text, vocab, sequence_to_tensor from earlier cells) ---
            
            # sequence_to_tensor expects a list of texts
            input_tensor = sequence_to_tensor([cleaned_text], vocab, MAX_SEQ_LENGTH).to(device)
            
            # 2. Inference
            logits = model(input_tensor)
            
        elif model_name == 'BERT':
            # --- BERT-based Preprocessing (Uses live_tokenizer) ---
            if live_tokenizer is None:
                return "BERT Tokenizer Error", 0.0
                
            encoding = live_tokenizer.encode_plus(
                cleaned_text,
                add_special_tokens=True,
                max_length=MAX_SEQ_LENGTH,
                padding='max_length',
                truncation=True,
                return_attention_mask=True,
                return_tensors='pt', # Returns PyTorch tensors
            )
            
            # Move inputs to the same device as the model
            input_ids = encoding['input_ids'].to(device)
            attention_mask = encoding['attention_mask'].to(device)
            
            # 2. Inference (BERT returns an object with .logits)
            outputs = model(input_ids, attention_mask=attention_mask)
            logits = outputs.logits
            
        else:
            return "Invalid Model Name", 0.0

    # 3. Output Calculation
    probabilities = F.softmax(logits, dim=1) # Get probabilities over the two classes
    confidence, predicted_index = torch.max(probabilities, dim=1)
    
    # 4. Format Output
    sentiment_label = "Positive (1)" if predicted_index.item() == 1 else "Negative (0)"
    confidence_score = confidence.item() * 100
    
    return sentiment_label, confidence_score
```

```python
# =========================================================================
# CELL 11: Interactive Application Demo
# =========================================================================

def run_sentiment_app():
    """Starts a simple command-line interface for sentiment prediction."""
    print("==============================================")
    print("   Real-time Sentiment Predictor (4 Models)   ")
    print("==============================================")
    print(f"Models Available: {list(loaded_models.keys())} (Type 'exit' to quit)")
    print("-" * 50)
    
    while True:
        try:
            # Get user input for the review
            review = input("\nEnter a movie review (or 'exit'): \n> ")
            
            if review.lower() == 'exit':
                print("\nExiting application. Goodbye!")
                break
            
            if not review.strip():
                print("Please enter a non-empty review.")
                continue

            print("\n--- Model Predictions ---")
            
            # List of models to run against the review
            model_list = ['RNN', 'LSTM', 'GRU', 'BERT']
            comparison_results = []
            
            for name in model_list:
                sentiment, confidence = predict_sentiment_live(review, name)
                
                # Format output for the table
                comparison_results.append({
                    'Model': name,
                    'Predicted Sentiment': sentiment,
                    'Confidence': f'{confidence:.2f}%'
                })

            # Display results in a clean DataFrame table
            comparison_df = pd.DataFrame(comparison_results)
            print(comparison_df.to_markdown(index=False))
            print("-" * 50)

        except KeyboardInterrupt:
            print("\nExiting application. Goodbye!")
            break
        except Exception as e:
            print(f"\nAn error occurred: {e}. Shutting down demo.")
            break

# Execute the application
run_sentiment_app()
```
